```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = "/content/Wholesale customers data.csv"
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
print(df.info())
```

```python
df.drop(['Channel','Region'], axis=1,inplace=True)
```

```python
array=df.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan=DBSCAN(eps=0.7, min_samples=8)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
dbscan=DBSCAN(eps=1.0, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
dbscan=DBSCAN(eps=1.8, min_samples=7)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
c1 = pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
pd.concat([df,c1],axis=1)
```

```python
fig = plt.figure(figsize=(10,6))
```

```python
ax = fig.add_subplot(111, projection='3d')
```

```python
ax.scatter(df['Fresh'],df['Milk'],df['Grocery'], c=df['cluster'],cmap='table',alpha=0.7)
ax.set_xlabel("Fresh")
ax.set_ylabel("Milk")
ax.set_zlabel("Grocery")
```

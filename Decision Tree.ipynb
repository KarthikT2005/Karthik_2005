```python
"""Decision Tree using C5.0 (Entropy) and CART (Gini)"""

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing

file_name = 'iris.csv'
iris = pd.read_csv(file_name)

print(iris.head())

label_encoder = preprocessing.LabelEncoder()
iris['Species'] = label_encoder.fit_transform(iris['Species'])

X = iris.iloc[:, 0:4]
y = iris['Species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
class_names = ['setosa', 'versicolor', 'virginica']

"""### Building Decision Tree Classifier using Entropy (C5.0)"""

model_entropy = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model_entropy.fit(X_train, y_train)

plt.figure(figsize=(8, 6))
tree.plot_tree(model_entropy, feature_names=feature_names, class_names=class_names, filled=True)
plt.show()

preds_entropy = model_entropy.predict(X_test)
print("Predictions using Entropy:", preds_entropy)

accuracy_entropy = np.mean(preds_entropy == y_test)
print("Accuracy (Entropy):", accuracy_entropy)

"""### Building Decision Tree Classifier using Gini (CART)"""

model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
model_gini.fit(X_train, y_train)


preds_gini = model_gini.predict(X_test)
print("Predictions using Gini:", preds_gini)

accuracy_gini = np.mean(preds_gini == y_test)
print("Accuracy (Gini):", accuracy_gini)

"""### Decision Tree Regression Example"""

X_reg = iris.iloc[:, 0:3].values
y_reg = iris.iloc[:, 3].values

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.33, random_state=1)
model_regressor = DecisionTreeRegressor()
model_regressor.fit(X_train_reg, y_train_reg)
regressor_score = model_regressor.score(X_test_reg, y_test_reg)
print("Decision Tree Regression R^2 Score:", regressor_score)
```
